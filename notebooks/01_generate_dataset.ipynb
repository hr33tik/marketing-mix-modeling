```python
import sys
from pathlib import Path

sys.path.append(str(Path("../src").resolve()))
```

```python
from dataset_generator import (
    generate_state_week_grid,
    generate_channel_activity,
    generate_calendar_features,
    generate_true_media_effects,
    generate_sales,
    generate_weekly_spends
)
```

```python
df = generate_state_week_grid()

df = generate_channel_activity(df)

df = generate_calendar_features(df)

media_df = generate_true_media_effects(df)

df = generate_sales(df, media_df)

weekly_spends = generate_weekly_spends(df)
```

```python
from pathlib import Path

output_folder = Path("../data/raw")

output_folder.mkdir(
    parents=True,
    exist_ok=True
)

df.to_csv(
    output_folder / "raw_data.csv",
    index=False
)

weekly_spends.to_csv(
    output_folder / "weekly_spends.csv",
    index=False
)

print("Datasets saved successfully!")
```
